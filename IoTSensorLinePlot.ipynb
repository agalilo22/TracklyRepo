```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Week 7: Line Plot of IoT Sensor Readings Over Time ---
print("\n--- Starting Week 7: Line Plot of IoT Sensor Readings Over Time ---")

# Access your preprocessed data from Homework No. 3 (which is cleaned_iot_data.csv)
# Load cleaned IoT data
try:
    df = pd.read_csv("cleaned_iot_data.csv")
    print("✅ Cleaned IoT data loaded successfully!")
except FileNotFoundError:
    print("❌ Error: 'cleaned_iot_data.csv' not found. Please ensure Week 6 was run successfully and generated the file.")
    exit()

# Display first few rows to verify data
print("\nPreview of loaded cleaned_iot_data.csv:")
print(df.head())

# To ensure proper plotting, convert the timestamp column into a datetime format.
# This is crucial for time-series plotting.
df["timestamp"] = pd.to_datetime(df["timestamp"])
print("\n'timestamp' column converted to datetime format.")
print(df.info()) # Verify data types, especially for 'timestamp'

# --- OPTIONAL: Filter out non-numeric data types for a cleaner plot ---
# If your 'data_type' column contains entries like "Status" which always result in
# a 'numeric_value' of 0 and you don't want them on the line graph,
# uncomment and use the following lines.
# This will make the plot focus only on truly numerical sensor readings.
#
# types_to_plot = ['Temperature', 'Location_Lat', 'Location_Lon'] # Adjust as per your data types
# df_plot = df[df['data_type'].isin(types_to_plot)].copy()
# print(f"\nFiltered data for plotting. Only including: {types_to_plot}")
# if df_plot.empty:
#     print("Warning: Filtered DataFrame is empty. No data to plot.")
#     exit()
# else:
#     # Use df_plot for plotting instead of df
#     pass
#
# If you comment out the filtering, the original 'df' will be used for plotting.
df_plot = df.copy() # By default, use the full DataFrame for plotting

# Set the visualization style for better aesthetics
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(14, 7))  # Adjust figure size for better readability, slightly larger
# Use hue="data_type" to get separate lines for different sensor types
# Ensure 'numeric_value' is suitable for plotting (i.e., not all zeros if that's not intended)
# Use df_plot if you implemented the optional filtering above, otherwise use df
sns.lineplot(x="timestamp", y="numeric_value", hue="data_type", marker="o", data=df_plot) # Changed data=df_plot

# Rotate x-axis labels for better readability, especially with many timestamps
plt.xticks(rotation=45, ha='right') # 'ha' (horizontal alignment) helps with rotation overlap

# Add title and labels
plt.title("IoT Sensor Readings Over Time by Data Type", fontsize=16) # More descriptive title
plt.xlabel("Timestamp", fontsize=13)
plt.ylabel("Sensor Value (Numeric)", fontsize=13)

# Add grid lines for better readability on both axes
plt.grid(True, linestyle='--', alpha=0.7)

# Show legend for sensor types
plt.legend(title="Sensor Type", bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside plot
plt.tight_layout() # Adjust layout to prevent labels from overlapping

# Display the plot
print("\nDisplaying the line plot...")
plt.show()

print("\n--- Week 7: Line Plotting Complete ---")

# --- Reminder for GitHub Upload ---
print("\n--- Reminder for GitHub Upload ---")
print("1. Save this updated Python script (.py) or Jupyter Notebook (.ipynb).")
print("2. If you want to save the plot image:")
print("   - You can add `plt.savefig('iot_sensor_readings_plot.png')` before `plt.show()`.")
print("     (Make sure to run the script once to generate the plot, then uncomment and run again to save.)")
print("3. Navigate to your GitHub repository in your web browser.")
print("4. Click 'Add file' -> 'Upload files'.")
print("5. Drag and drop or select your updated script/notebook and any saved plot images.")
print("6. Add a concise commit message (e.g., 'Feat: Week 7 IoT Data Visualization').")
print("7. Click 'Commit changes' to save your work to the repository.")
```
